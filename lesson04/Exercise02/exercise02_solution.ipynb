## Religious demographics

```python
# Import statements
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

Use pandas to read the data located in the subfolder data.

```python
# Load dataset
data = pd.read_csv('./data/religions.csv')
```

Use a pie chart to visualize the religious demographics. Highlight one religion of your choice using the explode parameter. Show the percentages for each slice and add a title.

```python
# Create figure
plt.figure(figsize=(8, 7), dpi=300)
# Create pie plot
plt.pie('Percentage', explode=(0, 0, 0.1, 0, 0, 0, 0), labels='Religion', data=data, autopct='%1.1f%%')
# Add title
plt.title('Religious demographics')
# Show plot
plt.show()
```
